```python
plt.figure(figsize = (11,6))

plt.hist(data=data, x="fixed acidity", color="orange",edgecolor="linen",alpha=0.5,bins=5)

plt.title("history of fixed acidity")
plt.xlabel("fixed acidity")
plt.ylabel("count")
plt.show()
```

```python
# caculate the mean of fixed acidity feactures 

data["fixed acidity"].mean()
```

```python
# caculate the median of fixed acidity feactures 

data["fixed acidity"].median()
```

```python
# create histogram of fixed "fixed acidity" feacture and also how to mean and median 

plt.figure(figsize = (11,6))

sns.histplot(data = data, x="fixed acidity", color="orange",edgecolor = "linen", alpha=0.5, bins=5)

plt.title("history of fixed acidity")
plt.xlabel("fixed acidity")
plt.ylabel("count")
plt.vlines(data["fixed acidity"].mean(), ymin=0, ymax=4000, colors="blue", label="mean")
plt.vlines(data["fixed acidity"].median(), ymin=0, ymax=4000, colors="red", label="median")

plt.legend()
plt.show()
```

```python
# create histogram of the "volatile acidity" ,acidity feacture

plt.figure (figsize=(11,6))

sns.histplot(data=data, x="volatile acidity", color="green", edgecolor="linen", alpha=0.5, bins=5)

plt.title("histogram of volatile acidity")
plt.xlabel("volitile acidity")
plt.ylabel("count")

plt.show()
```

```python
# create distplot of the "volatile acidity" ,acidity feacture

plt.figure (figsize=(11,6))

sns.distplot(data["volatile acidity"], color="blue")

plt.title("histogram of volatile acidity")
plt.xlabel("volitile acidity")
plt.ylabel("count")

plt.show()
```

```python
# show skewness of volatile acidity

data["volatile acidity"].skew()
```

```python
# calculate the mean "volitile acidity" feacture

data["volatile acidity"].mean()
```

```python
# calculate the median "volitile acidity" feacture

data["volatile acidity"].median()
```

```python
# create histogram of "volatile acidity" feacture and show the mean and median

plt.figure (figsize = (11,5))

sns.histplot(data=data, x="volatile acidity", color="yellow",edgecolor = "linen", alpha=0.5, bins=5)

plt.title("histogram of volatile acidity")
plt.xlabel("volatile acidity")
plt.ylabel("dencity")
plt.vlines(data["volatile acidity"].mean(), ymin=0, ymax=4000, colors="blue", label="mean")
plt.vlines(data["volatile acidity"].median(), ymin=0, ymax=4000, colors="red", label="median")


plt.show()
```

```python
# create a histogram of "citric acid " feacture

plt.figure(figsize = (11,5))

sns.histplot(data=data, x="citric acid", color="yellow",edgecolor = "linen", alpha=0.5, bins=5)

plt.title("histogram of citric acid")
plt.xlabel("citric acid")
plt.ylabel("count")
plt.show()
```

```python
# calculate the mean feacture "citric acid" feacture 

data["citric acid"].mean()
```

```python
# calculate the median feacture "citric acid" feacture 

data["citric acid"].median()
```

```python
# create a histogram of the "citric acid" feacture 

plt.figure(figsize = (11,6))

sns.histplot(data=data, x="citric acid", color="yellow",edgecolor = "linen", alpha=0.5, bins=5)

plt.vlines(data["citric acid"].mean(), ymin=0, ymax=4000, colors="blue", label="mean")
plt.vlines(data["citric acid"].median(), ymin=0, ymax=4000, colors="blue", label="median")

plt.title("histogram of citric acid")
plt.xlabel("citric acid")
plt.ylabel("count")
plt.show()
```

```python
# calculate distplot using "citric acid" feacture 

plt.figure(figsize = (11,5))

sns.distplot(data["citric acid"], color="blue")

plt.title("hist plot of citric acid")
plt.xlabel("citric acid")
plt.ylabel("density")

plt.show()
```

```python
# create a count plot of the 'quality' feacture

plt.figure(figsize=(11,6))
sns.countplot(x=data["quality"])
plt.title("quality")
plt.xlabel("quality")
plt.ylabel("count")
plt.show()
```
